```python
from transformers import pipeline
from IPython.display import FileLink

# Load Seamless M4T model on GPU
pipe = pipeline(
    "automatic-speech-recognition",
    model="facebook/seamless-m4t-v2-large",
    device=0,                  # GPU (0 = first CUDA device)
    torch_dtype="float16"      # faster + less VRAM
)
```

```python
# Input video/audio file
AUDIO = "/kaggle/input/datasetfacebookseamless-m4t-v2-large/OnnoRokom Pathshala.mp4"
```

```python
# Run transcription
print("ðŸ”„ Running transcription with facebook/mms-1b-all...")
result = pipe(
    AUDIO,
    chunk_length_s=15,   # safe chunk size for Kaggle T4
    batch_size=2,        # balanced for 2 GPUs
    generate_kwargs={"tgt_lang": "ben"}   # âœ… Bangla target language
)
```

```python
# Save transcript
transcript = result["text"]
with open("transcript_mms_bangla.txt", "w", encoding="utf-8") as f:
    f.write(transcript)

print("âœ… Done! Transcript saved to transcript_mms_bangla.txt")
print("="*50)
print(transcript[:400] + "..." if len(transcript) > 400 else transcript)
```
